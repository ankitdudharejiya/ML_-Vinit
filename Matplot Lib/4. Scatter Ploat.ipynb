# Scatter plots are used to observe relationship between variables and uses dots to represent the relationship between them. The scatter() method in the matplotlib library is used to draw a scatter plot. Scatter plots are widely used to represent relation among variables and how change in one affects the other. 
Syntax 
The syntax for scatter() method is given below:

# matplotlib.pyplot.scatter(x_axis_data, y_axis_data, s=None, c=None, marker=None, cmap=None, vmin=None, vmax=None, alpha=None, linewidths=None, edgecolors=None) 
 

# Scatter plots are used to plot data points on horizontal and vertical axis in the attempt to show how much one variable is affected by another. Each row in the data table is represented by a marker the position depends on its values in the columns set on the X and Y axes. A third variable can be set to correspond to the color or size of the markers, thus adding yet another dimension to the plot.

```python
import matplotlib.pyplot as plt # For visualization
import pandas as pd # for data cleaing & analysis
```

```python
# Dataset taken from kaggel.com
df_google_play_store_apps = pd.read_csv("googleplaystore.csv")
df_google_play_store_apps.shape # Get shape of dataset in rows and colms
```

```python
# want to use small datas
# Dataset taken from kaggel.com
df_google_play_store_apps = pd.read_csv("googleplaystore.csv" , nrows=100
                                       )
df_google_play_store_apps.shape # Get shape of dataset in rows and colms
```

```python
df_google_play_store_apps 
```

```python
#  Rating as X and Review as Y

x= df_google_play_store_apps["Rating"]
y= df_google_play_store_apps["Reviews"]
```

```python
# Scatter Plot

plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
# Diffrent paramters 
```

```python
plt.scatter(x,y , color='y')
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize=[16,9])
plt.scatter(x,y , color='y')
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
#marker
plt.figure(figsize=[16,9])
plt.scatter(x,y , color='y' , marker = '*')
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
#marker size
plt.figure(figsize=[16,9])
plt.scatter(x,y , color='y' , marker = '*' , s =100)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
#opasity
plt.figure(figsize=[16,9])
plt.scatter(x,y , color='y' , marker = '*' , s =100 , alpha=0.1)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
#opasity
plt.figure(figsize=[16,9])
plt.scatter(x,y , color='y' , marker = '*' , s =100 , alpha=0.8)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
#Linewidth of starts
plt.figure(figsize=[16,9])
plt.scatter(x,y , color='y' , marker = '*' , s =100 , alpha=0.8 , linewidths=5)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
#Linewidth color
plt.figure(figsize=[16,9])
plt.scatter(x,y , color='y' , marker = '*' , s =100 , alpha=0.8 , linewidths=10 , edgecolors= 'g')
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
# verts parameter works like marker
plt.figure(figsize=[16,9])
plt.scatter(x,y , color='y' , marker = '*' , s =100 , alpha=0.8 , linewidths=10 , edgecolors= 'g')
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
#multiple
plt.figure(figsize=[16,9])
plt.scatter(x,y , color='y' , marker = '*' , s =100 , alpha=0.8 , linewidths=10 )
plt.scatter(x,df_google_play_store_apps['Installs'] , color='r' , marker = '*' , s =100 , alpha=0.8 , linewidths=10)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews & Installs")
plt.show()
```
