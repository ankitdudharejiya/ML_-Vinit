```python
df1= pd.DataFrame({'ID' : [1,2,3,4] , 'Name' : ['A','B','C','D'], 'Class': [5,6,7,8]})
df1
```

```python
df2= pd.DataFrame({'ID' : [3,4] , 'Name' : ['C','D'], 'Class': [7,8]})
df2
```

```python
pd.concat(objs= [df1,df2]) # row wise
```

```python
pd.concat(objs= [df1,df2] , axis=1) # column wise
```

```python
#print only common value - joint parameter


pd.concat(objs= [df1,df2] ,join='inner' ,axis=1)
```

```python
pd.concat(objs= [df1,df2] ,join='outer' ,axis=1)
```

```python
#key Parameter 
 
```

```python
df1= pd.DataFrame({'ID' : [1,2,3,4] , 'Name' : ['A','B','C','D'], 'Class': [5,6,7,8]})
df1
 
```

```python
df2= pd.DataFrame({'ID' : [5,6,7,8] , 'Name' : ['E','F','G','H'], 'Class': [9,10,11,12]})
df2
```

```python
pd.concat(objs= [df1,df2] , keys = ['df1','df2'])
```

```python
pd.concat(objs= [df1,df2] , keys = ['a1','a2'])
```

```python
pd.concat(objs= [df1,df2] ,axis=1, keys = ['a1','a2'])
```
