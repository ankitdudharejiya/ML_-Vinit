# Histogram Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_students_age = np.random.randint(18,45, (100))
py_students_age = np.random.randint(15,40, (100))
```

```python
ml_students_age
```

```python
py_students_age 
```

#  A histogram is an accurate representation of the distribution of numerical data. It is an estimate of the probability distribution of a continuous variable. It is a kind of bar graph.

```python
plt.hist(ml_students_age)
plt.show()
```

```python
# Title & lables
plt.hist(ml_students_age)
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.show()
```

```python
# bins - To define bar category range 
```

```python
bins = [15,20,25,30,35,40,45]
```

```python
plt.hist(ml_students_age , bins)
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.show()
```

# to show bar Proper


```python
plt.hist(ml_students_age , bins , rwidth=0.8) # defaul value is 1 means 100%
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.show()
```

```python
plt.hist(ml_students_age , bins , rwidth=0.8 , histtype= "step") 
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.show()
```

```python
# align paramter - to aligh bar postion
```

# orientation - to give horizontal or veritcal presentation of BAR

```python
plt.hist(ml_students_age , bins , rwidth=0.8 , histtype= "bar" , orientation='horizontal') 
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.show()
```

# to change Bar colour

```python
plt.hist(ml_students_age , bins , rwidth=0.8 , histtype= "bar" , orientation='horizontal', color='c') 
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.show()
```

# lable - for histogram

```python
plt.hist(ml_students_age , bins , rwidth=0.8 , histtype= "bar" , orientation='horizontal', label= 'ML Student') 
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.show()
```

```python
# we are not gettin any lable as we have not use LEGEND fucntion
```

```python
plt.hist(ml_students_age , bins , rwidth=0.8 , histtype= "bar" , orientation='horizontal', label= 'ML Student') 
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

# 2 Information in One Histogram

```python
plt.hist(ml_students_age , bins , rwidth=0.8 , histtype= "bar" , orientation='horizontal', label= 'ML Student') 
plt.hist(py_students_age , bins , rwidth=0.8 , histtype= "bar" , orientation='horizontal', label= 'Py Student') 
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

```python
plt.hist(ml_students_age , bins , rwidth=0.8 , histtype= "bar" , orientation='horizontal', label= 'ML Student') 
plt.hist(py_students_age , bins , rwidth=0.8 , histtype= "bar" , orientation='vertical', label= 'Py Student') 
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

```python
plt.hist(ml_students_age , bins , rwidth=0.8 , histtype= "bar" , orientation='vertical', label= 'ML Student') 
plt.hist(py_students_age , bins , rwidth=0.8 , histtype= "bar" , orientation='vertical', label= 'Py Student') 
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

# to plot seprate bar in multiple inuput

```python
plt.hist([ml_students_age , py_students_age] , bins , rwidth=0.8 , histtype= "bar" , orientation='vertical',
         color=['m', 'y' ],label= ['ML Student' , 'Py Student']) 
#plt.hist(py_students_age , bins , rwidth=0.8 , histtype= "bar" , orientation='vertical', label= 'Py Student') 
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

# increase Histogram size

```python
bins = [15,20,25,30,35,40,45]
plt.figure(figsize= (16,9))
plt.hist([ml_students_age , py_students_age] , bins , rwidth=0.8 , histtype= "bar" , orientation='vertical',
         color=['m', 'y' ],label= ['ML Student' , 'Py Student']) 
#plt.hist(py_students_age , bins , rwidth=0.8 , histtype= "bar" , orientation='vertical', label= 'Py Student') 
plt.title("ML Students age histograms", fontsize=15)
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```
