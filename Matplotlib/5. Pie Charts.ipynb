```python
import matplotlib.pyplot as plt
```

```python
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 
           'Data Sciece']
class1_students = [45, 15, 35, 25, 30]
```

```python
#Matplotlib Simple pie Chart
plt.pie([1]) # Plot pie chart of value [1]
plt.show() # To show Pie chart
```

```python
#Pie Plot

plt.pie(class1_students)
plt.show()
```

```python
# now there is no lable so we want to display it

plt.pie(class1_students, labels = classes)
plt.show()
```

```python
#now supppose we want to split some part from chart then we use EXPLODE paramater
explode = [0,0,0,0,0] # To slice the perticuler section
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 
           'Data Sciece']
class1_students = [45, 15, 35, 25, 30]

plt.pie(class1_students, labels = classes , explode=explode)
plt.show()
```

```python
explode = [0,0,0.2,0,0] # To slice the perticuler section
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 
           'Data Sciece']
class1_students = [45, 15, 35, 25, 30]
colors = ['c','r','y','b','k']

plt.pie(class1_students, labels = classes , explode=explode , colors=colors)
plt.show()
```

```python
# autopct = To show value in Percentage format
```

```python
explode = [0.03,0,0.1,0,0] # To slice the perticuler section
colors = ["c", 'b','r','y','g'] # Color of each section
textprops = {"fontsize":25} # Font size of text in pie chart
 
plt.pie(class1_students, # Values
        labels = classes, # Labels for each sections
        explode = explode, # To slice the perticuler section
        colors =colors, # Color of each section
        autopct = "%0.3f%%")# Show data in persentage for with 2 decimal point
        
plt.show() # To show pie chart only
```

```python
# shadow
explode = [0.03,0,0.1,0,0] # To slice the perticuler section
colors = ["c", 'b','r','y','g'] # Color of each section
textprops = {"fontsize":15} # Font size of text in pie chart
 
plt.pie(class1_students, # Values
        labels = classes, # Labels for each sections
        explode = explode, # To slice the perticuler section
        colors =colors, # Color of each section
        autopct = "%0.2f%%", # Show data in persentage for with 2 decimal point
        shadow = True) # Showing shadow of pie chart
plt.show()
```

```python
# to increase size of chart radius paramter

explode = [0.03,0,0.1,0,0] # To slice the perticuler section
colors = ["c", 'b','r','y','g'] # Color of each section
 
plt.pie(class1_students, # Values
        labels = classes, # Labels for each sections
        explode = explode, # To slice the perticuler section
        colors =colors, # Color of each section
        autopct = "%0.2f%%", # Show data in persentage for with 2 decimal point
        shadow = True, # Showing shadow of pie chart
        radius = 1.8) # Radius to increase or decrease the size of pie chart
plt.show()
```

```python
#if we want to plot accoordingly our requirement so we have to give angle  so will use STARTANGLE parameter
```

```python

explode = [0.03,0,0.1,0,0] # To slice the perticuler section
colors = ["c", 'b','r','y','g'] # Color of each section

plt.pie(class1_students, # Values
        labels = classes, # Labels for each sections
        explode = explode, # To slice the perticuler section
        colors =colors, # Color of each section
        autopct = "%0.2f%%", # Show data in persentage for with 2 decimal point
        shadow = True, # Showing shadow of pie chart
        radius = 1.4, # Radius to increase or decrease the size of pie chart 
       startangle = 60) # Start angle of first section
plt.show()
```

```python
# to plot pie chart in center

explode = [0.1,0,0.1,0,0] # To slice the perticuler section
colors = ["c", 'b','r','y','g'] # Color of each section
textprops = {"fontsize":15} # Font size of text in pie chart
 
plt.pie(class1_students, # Values
        labels = classes, # Labels for each sections
        explode = explode, # To slice the perticuler section
        colors =colors, # Color of each section
        autopct = "%0.2f%%", # Show data in persentage for with 2 decimal point
        shadow = True, # Showing shadow of pie chart
        radius = 2.4, # Radius to increase or decrease the size of pie chart 
       startangle = 90, # Start angle of first section
        textprops =textprops,
       center = (5,3)) 
 
plt.show() # To show pie chart only
```

```python
# to insert frame

explode = [0.1,0,0.1,0,0] # To slice the perticuler section
colors = ["c", 'b','r','y','g'] # Color of each section
textprops = {"fontsize":15} # Font size of text in pie chart
 
plt.pie(class1_students, # Values
        labels = classes, # Labels for each sections
        explode = explode, # To slice the perticuler section
        colors =colors, # Color of each section
        autopct = "%0.2f%%", # Show data in persentage for with 2 decimal point
        shadow = True, # Showing shadow of pie chart
        radius = 2.4, # Radius to increase or decrease the size of pie chart 
       startangle = 90, # Start angle of first section
        textprops =textprops,
       center = (2,3),
       frame = True) 
 
plt.show() # To show pie chart only
```

```python

wedgeprops = {"linewidth": 4, 'width':1, "edgecolor":"k"} # Width = 1
plt.pie(
        class1_students, 
        labels = classes, 
        explode = explode, 
        colors = colors, 
        autopct = "%0.2f%%", 
        pctdistance = 0.6, 
        shadow =True, 
        labeldistance = 1.8, 
        startangle = 270,
        radius = 1, 
        counterclock = True, 
        wedgeprops = wedgeprops,
        textprops = textprops,
        center=(2, 3),
        frame=True,
        rotatelabels=True
        ) 
plt.show()
```

```python
# to plot legend

explode = [0.1,0,0.1,0,0] # To slice the perticuler section
colors = ["c", 'b','r','y','g'] # Color of each section
textprops = {"fontsize":15} # Font size of text in pie chart
 
plt.pie(class1_students, # Values
        labels = classes, # Labels for each sections
        explode = explode, # To slice the perticuler section
        colors =colors, # Color of each section
        autopct = "%0.2f%%", # Show data in persentage for with 2 decimal point
        shadow = True, # Showing shadow of pie chart
        radius = 2.4, # Radius to increase or decrease the size of pie chart 
       startangle = 90, # Start angle of first section
        textprops =textprops) 
plt.legend(loc=10)
 
plt.show() # To show pie chart only
```
