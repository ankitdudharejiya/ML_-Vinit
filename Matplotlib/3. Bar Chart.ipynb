
# Ploting Bar Chart

Plot Bar Chart with a different way
To plot bar chart using the matplotlib python library, use plt.bar() or plt.barh() methods.

Syntax: plt.bar(
                             x,
                             height,
                             width=0.8,
                             bottom=None,
                             *,
                             align=center,
                             data=None,
                             **kwargs,
                             )

Parameters
-
x : sequence of scalars
height : scalar or sequence of scalars
width : scalar or array-like, optional .(default: 0.8).
bottom : scalar or array-like, optional
align : {center, edge}, optional, .default: center


Other Parameters
-
color : scalar or array-like, optional
edgecolor : scalar or array-like, optional
linewidth : scalar or array-like, optional
tick_label : string or array-like, optional . name of Bar
xerr, yerr : scalar or array-like of shape(N,) or shape(2,N), optional
ecolor : scalar or array-like, optional, default: black
capsize : scalar, optional
error_kw : dict, optional
log : bool, optional, default: False
orientation : {vertical, horizontal}, optional


See also

barh: Plot a horizontal bar plot.

Other optional kwargs:

agg_filter: a filter function, which takes a (m, n, 3) float array and a dpi value, and returns a (m, n, 3) array
===== alpha: float or None
animated: bool
antialiased: unknown
capstyle: {butt, round, projecting}
clip_box: `.Bbox`
clip_on: bool
clip_path: [(`~matplotlib.path.Path`, `.Transform`) | `.Patch` | None]
===== color: color
contains: callable
===== edgecolor: color or None or auto
===== facecolor: color or None
===== figure: `.Figure`
fill: bool
gid: str
hatch: {/, \\, |, -, +, x, o, O, ., *}
in_layout: bool
joinstyle: {miter, round, bevel}
===== label: object
===== linestyle: {-, , -., :, , (offset, on-off-seq), }
===== linewidth: float or None for default
path_effects: `.AbstractPathEffect`
picker: None or bool or float or callable
rasterized: bool or None
sketch_params: (scale: float, length: float, randomness: float)
snap: bool or None
transform: `.Transform`
url: str
===== visible: bool
zorder: float

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
# Dataset of 'Sofcon Artificial Intelligence Class"
 
#classes = ["Python", "R", "Artificial Intelligence", "Machine Learning", "Data Science"]
 
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10] # out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

```python
plt.bar(classes, class1_students) # Plot vertical Bar Chart
```

# Horizontal Ploting

```python
plt.barh(classes, class1_students) # Plot horizontal Bar Chart
plt.show()
```

# Diffrent Parameters in BAR function

```python
#assign Bar Width

plt.bar(classes, class1_students, width=0.2)
```

```python
# Allign Parameter by default Center

plt.bar(classes, class1_students, width=0.2 , align= 'edge')
```

```python
# Color Paramtere 

plt.bar(classes, class1_students, width=0.2 , align= 'edge' , color='y')
```

```python
# bar edge color change

plt.bar(classes, class1_students, width=0.2 , align= 'edge' , color='y' , edgecolor='k')
```

```python
# Change Bar Line Width

plt.bar(classes, class1_students, width=0.2 , align= 'edge' , color='y' , edgecolor='k' , linewidth = 2)
```

```python
# to change opacity of Bar will use alpha parameter

plt.bar(classes, class1_students, width=0.2 , align= 'edge' , color='y' , edgecolor='k' , 
        linewidth = 2 , alpha= 0.5) # alpha value between 0 to 1
```

```python
 # To change line style of bar
plt.bar(classes, class1_students, width=0.2 , align= 'edge' , color='y' , edgecolor='k' , 
        linewidth = 2 , alpha= 0.5 , linestyle='--')
    
```

```python
 # To change line style of bar
plt.bar(classes, class1_students, width=0.2 , align= 'edge' , color='y' , edgecolor='k' , 
        linewidth = 2 , alpha= 0.5 , linestyle=':')
    
```

```python
# To assign lable
 
plt.bar(classes, class1_students, width=0.2 , align= 'edge' , color='y' , edgecolor='k' , 
        linewidth = 2 , alpha= 0.5 , label ='Class 1 Students')
plt.legend()
```

```python
# To hide graphs

plt.bar(classes, class1_students, width=0.2 , align= 'edge' , color='y' , edgecolor='k' , 
        linewidth = 2 , alpha= 0.5 , label ='Class 1 Students' , visible=False)
```

```python
# Grid 
style.use('ggplot')
plt.bar(classes, class1_students, width=0.2 , align= 'edge' , color='y' , edgecolor='k' , 
        linewidth = 2  , label ='Class 1 Students')


```

```python
# To increase bar size will use FIGURE function


plt.figure(figsize=[16,9])
plt.bar(classes, class1_students, width=0.2 , align= 'edge' , color='y' , edgecolor='k' , 
        linewidth = 2  , label ='Class 1 Students')
```

```python
# if we use once style then we dont need to use it every time 
```

```python
# Horizontal Plot
plt.figure(figsize=[16,9])
plt.barh(classes, class1_students, width=0.2 , align= 'edge' , color='y' , edgecolor='k' , 
        linewidth = 2  , label ='Class 1 Students' )
```

Note : see above error that we can not use width with barh

```python
# Horizontal Plot
plt.figure(figsize=[16,9])
plt.barh(classes, class1_students , align= 'edge' , color='y' , edgecolor='k' , 
        linewidth = 2  , label ='Class 1 Students' )
```

```python
# X - Y lable and title assign

plt.figure(figsize=[16,9])
plt.bar(classes, class1_students, width=0.2 , align= 'edge' , color='y' , edgecolor='k' , 
        linewidth = 2  , label ='Class 1 Students')
plt.title("Bar Chart of Sofcon Class", fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("No. of Students", fontsize = 15)
plt.show()
```

# Multiple Graphs

```python
plt.figure(figsize=(16,9))
 
plt.bar(classes, class1_students, width = 0.2, color = "b",
        label =" Class 1 Students") #visible=False
 
plt.bar(classes, class2_students, width = 0.2, color = "g",
        label =" Class 2 Students" ) 
 
plt.title("Bar Chart of Sofcon Class", fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("No. of Students", fontsize = 15)
plt.show()
```

# Our class value are not displayes so we can display using below function (xtics)

```python
plt.figure(figsize=(16,9))
 
classes_index = np.arange(len(classes))
 
width = 0.2
 
plt.bar(classes_index, class1_students, width , color = "b",
        label =" Class 1 Students") #visible=False
 
plt.bar(classes_index + width, class2_students, width , color = "g",
        label =" Class 2 Students") 
 
plt.bar(classes_index + width + width, class3_students, width , color = "y",
        label =" Class 3 Students") 
 
plt.xticks(classes_index + width, classes, rotation = 20)
plt.title("Bar Chart of Sofcon Class", fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("No. of Students", fontsize = 15)
plt.show()
```
